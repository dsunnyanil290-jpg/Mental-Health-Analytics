**VISUALISATION**

**Mental Health Risk Distribution**

```python
plt.figure(figsize=(8,5))

sns.countplot(
    data=df,
    x='mental_health_risk',
    palette='Set2'
)

plt.title('Mental Health Risk Distribution',
          fontsize=16,
          fontweight='bold')

plt.xlabel('Risk Category')
plt.ylabel('Count')

plt.show()
```

**Average Stress by Gender**

```python
plt.figure(figsize=(8,5))

sns.barplot(
    data=df,
    x='gender',
    y='stress_level',
    palette='viridis'
)

plt.title('Average Stress Level by Gender',
          fontsize=16,
          fontweight='bold')

plt.xlabel('Gender')
plt.ylabel('Average Stress Level')

plt.show()
```

**Depression Score by Gender (Box Plot)**

```python
plt.figure(figsize=(8,5))

sns.boxplot(
    data=df,
    x='gender',
    y='depression_score',
    palette='pastel'
)

plt.title('Depression Score by Gender',
          fontsize=16,
          fontweight='bold')

plt.xlabel('Gender')
plt.ylabel('Depression Score')

plt.show()
```

**Social Support vs Anxiety**

```python
plt.figure(figsize=(8,5))

sns.regplot(
    data=df,
    x='social_support_score',
    y='anxiety_score',
    scatter_kws={'alpha':0.5},
    line_kws={'color':'red'}
)

plt.title('Social Support vs Anxiety',
          fontsize=16,
          fontweight='bold')

plt.xlabel('Social Support Score')
plt.ylabel('Anxiety Score')

plt.show()
```

**Productivity vs Stress Correlation**

```python
sns.regplot(
    data=df,
    x='stress_level',
    y='productivity_score'
)
```

**Treatment Seeking Behavior**

```python
plt.figure(figsize=(7,5))

sns.countplot(
    data=df,
    x='seeks_treatment',
    hue='seeks_treatment',
    palette='Set2',
    legend=False
)

plt.title('Treatment Seeking Behavior',fontsize=16,fontweight='bold')
plt.xlabel('Seeking Treatment')
plt.ylabel('Count')

plt.show()
```

**Depression by Age Group**

```python
df['age_group'] = pd.cut(
    df['age'],
    bins=[18,25,35,45,55,65],
    labels=['18-25','26-35','36-45','46-55','56-65']
)
```

```python
plt.figure(figsize=(8,5))

sns.boxplot(
    data=df,
    x='age_group',
    y='depression_score',
    palette='pastel'
)

plt.title('Depression Score by Age Group',fontsize=16,fontweight='bold')
plt.xlabel('Age Group')
plt.ylabel('Depression Score')

plt.show()
```

**Heatmap**

```python

plt.figure(figsize=(10,6))
sns.heatmap(
    df.corr(numeric_only=True),
    annot=True,
    cmap='coolwarm'
)
plt.show()
```
